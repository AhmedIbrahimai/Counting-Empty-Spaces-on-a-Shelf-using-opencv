```python
import cv2
import pickle
import numpy as np
import cvzone

cap = cv2.VideoCapture('sf3.mp4')

# Define the output video file
output_file = 'output_video.avi'

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # You can change the codec as needed
fps = 20  # Frames per second
frame_size = (1020, 800)  # Size of the frames
out = cv2.VideoWriter(output_file, fourcc, fps, frame_size)

with open('shelf', 'rb') as f:
    points = pickle.load(f)

w, h = 66, 189

def crop(fa):
    counter = 0
    for pts in points:
        x, y = pts
        crop = fa[y:y+h, x:x+w]
        cv2.imshow(str(x*y), crop)
        count = cv2.countNonZero(crop)
        cv2.putText(frame, str(count), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
        if count > 100:
            cv2.rectangle(frame, pts, (pts[0]+w, pts[1]+h), (0, 255, 0), 2)
        else:
            cv2.rectangle(frame, pts, (pts[0]+w, pts[1]+h), (0, 0, 255), 2)
            counter += 1
        cvzone.putTextRect(frame, f'Space count: {counter}', (50, 60), 2, 2)

while True:
    ret, frame = cap.read()
    if not ret:
        cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue

    frame = cv2.resize(frame, frame_size)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 1)
    thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 105, 9)
    median_blur = cv2.medianBlur(thresh, 5)
    kernel = np.ones((1, 1), np.uint8)
    dilated = cv2.dilate(median_blur, kernel, iterations=1)
    crop(dilated)

    # Write the frame to the output video
    out.write(frame)

    cv2.imshow("Image", frame)

    if cv2.waitKey(30) & 0xFF == 27:
        break

# Release the VideoWriter and close the windows
out.release()
cap.release()
cv2.destroyAllWindows()
```
